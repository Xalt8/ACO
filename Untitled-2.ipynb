```python
def plot(graph:nx.Graph, short_path:list[tuple])-> plt.Axes:
    
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(22,9))
    
    _pos = {node: (graph.nodes[node]['x'], graph.nodes[node]['y']) for node in graph.nodes}
    
    # Optimising graph
    nx.draw_networkx_nodes(graph, pos=_pos, node_size=400, ax=ax1, node_color='skyblue')
    nx.draw_networkx_labels(graph, pos=_pos, ax=ax1)
    
    #Get phermone levels
    phermone_levels = [max(graph[node1][node2]['phermone'],0.005)  for node1, node2 in G.edges()]

    nx.draw_networkx_edges(graph, edgelist=list(graph.edges()), ax=ax1, pos=_pos, width=phermone_levels, edge_color='r', style='-')
    ax1.set_xlim(-10, 100)
    ax1.set_ylim(0,110)
    ax1.set_title('Path found')
    ax1.axis("on")

    # Shortest path graph
    nx.draw_networkx_nodes(graph, pos=_pos, node_size=400, ax=ax2, node_color='skyblue')
    nx.draw_networkx_labels(graph, pos=_pos, ax=ax2)
    nx.draw_networkx_edges(graph, edgelist=short_path, ax=ax2, pos=_pos, width=2, edge_color='r', style='--')
    ax2.set_xlim(-10, 100)
    ax2.set_ylim(0,110)
    ax2.axis("on")
    ax2.set_title('Shortest path')
    # plt.close()
    plt.show()
    
    
```

```python
def random_back(pos, vel):
    inds = list(np.arange(pos.size))
    new_pos = pos + vel
    # print(new_pos)
    for j, n in enumerate(new_pos):
        if n in inds:
           inds.remove(n)
        else:
            closest_val = inds[np.argmin([np.abs(n-val) for val in inds])]
            new_pos[j] = closest_val
            inds.remove(closest_val)

    return new_pos

```

```python
def random_back2(pos, vel):
    inds = np.arange(pos.size)
    new_pos = pos + vel
    # print(new_pos)
    for j, n in enumerate(new_pos):
        if n in inds:
           inds = np.delete(inds, np.where(n==inds)[0][0])
        else:
            closest_ind = np.argmin([np.abs(n-val) for val in inds])
            new_pos[j] = inds[closest_ind]
            inds = np.delete(inds, closest_ind)

    return new_pos
```

```python
from numba import int64, types, njit, typed, typeof, float64
from numba.experimental import jitclass
import typing

city_spec=[
    ('name', types.unicode_type),
    ('x', int64),
    ('y', int64)]

@jitclass(city_spec)
class City:
    def __init__(self, name:str, x:int, y:int) -> None:
        self.name = name
        self.x = x
        self.y = y
```

```python
london = City('LOndon', 100, 200)
paris = City('Paris', 200, 400)

cities = typed.List()
cities.append(london)
cities.append(paris)
```
