```python
def get_path_tuples(short_path_list:list) -> list[tuple]:
    ''' Returns a list of tuples for a list of coordinates'''
    path_tuples = [(short_path_list[i], short_path_list[i+1]) for i in range(len(short_path_list)-1)]
    path_tuples.append((short_path_list[0], short_path_list[-1])) # get the first and last one
    return path_tuples    
```

```python
def get_distance(node1:tuple, node2:tuple) -> float:
    '''Returns the distance between 2 nodes/coordinates'''
    return np.linalg.norm(np.array(node1) - np.array(node2))
```

```python
def create_graph(coords:list[tuple]) -> nx.Graph:
    ''' Takes a list of coordinates and returns a graph object'''
    graph = nx.Graph()
    node_labels = np.arange(1, len(coords) +1) # 1...n
    # graph.add_nodes_from(node_labels)
    for node, (_x,_y) in zip(node_labels, coords):
        graph.add_node(node, x=_x, y=_y)
    return graph           
```

```python
def plot_graph(graph:nx.Graph) -> plt.Axes:
    fig, ax = plt.subplots(figsize=(14.562, 9))
    _pos = {node: (graph.nodes[node]['x'], graph.nodes[node]['y']) for node in graph.nodes}
    nx.draw(graph, pos=_pos, node_size=400, ax=ax, node_color='skyblue')
    nx.draw_networkx_labels(graph, pos=_pos)
    ax.set_xlim(0, 100)
    ax.set_ylim(0,110)
    plt.axis("on")
    ax.tick_params(left=True, bottom=True, labelleft=True, labelbottom=True)

    plt.show()

```

```python
def calculate_trip_distance(path:list[tuple], graph:nx.Graph) -> float:
    ''' Takes a list of node edges and returns the total 
        Euclidean distance of the trip '''
    total_distance = []
    for node1, node2 in path:
        total_distance.append(
            get_distance(
                (graph.nodes[node1]['x'], graph.nodes[node1]['y']), 
                (graph.nodes[node2]['x'], graph.nodes[node2]['y'])))
    return np.around(np.sum(total_distance), decimals=2)
```

### ACO algorithm steps
- Ant starts from a random location/vertex/edge <br>
- Moves along an edge to another vertex <br>
- Keeps a memory of nodes it has visited <br>
- Choose node: <br>
    - q = random value
    - If q < q0(0.5) -> max(city_score) -> exploitation
    - If q > q0(0.5) -> choose in proportion to city_score/sum_city_scores -> exploration

```python
from dataclasses import dataclass, field
import random
```

```python
@dataclass
class Ant:

    visited_nodes: list[int] = field(default_factory=list, init=False)
    tour: list[tuple] = field(default_factory=list, init=False)
    current_node:int = field(init=False)  
    graph: nx.Graph
    alpha = 1 # Importance of phermone
    beta = 1 # Importance of heuristics
    q0 = 0.5
    rho = 0.99 # used to decay the phermone trail

    
    def __post_init__(self):
        self.current_node = self.choose_start_node()
        self.visited_nodes.append(self.current_node)

    def choose_start_node(self) -> int:
        ''' Randomly chooses a node from a 
            list of nodes in a graph'''
        return random.choice(list(self.graph.nodes))
    
    def get_unvisited_nodes(self) -> list[int]:
        ''' Returns a list of unvisited nodes'''
        return [node for node in list(self.graph.nodes) if node not in self.visited_nodes]
    
    def score_node(self, node:int) -> float:
        ''' Scores the node based on phermone and distance'''
        return self.graph.nodes[self.current_node][node]['phermone']**self.alpha * \
                    (1/self.graph.nodes[self.current_node][node]['distance'])**self.beta

    # def exploitation_strategy(self, unvisited_nodes:list) -> int:
    #     ''' Returns a node after getting the max score of unvisited cities
    #         ==============================================================
    #         score_array: applies the score_node function to all unvisited 
    #         nodes and returns an array.
    #         argmax of the score array returns the index value of the max score
    #         Get the node from the unvisited node list using the index.  '''
       
    #     score_array = np.array(map(self.score_node, unvisited_nodes))
    #     return unvisited_nodes[np.argmax(score_array)]
        

    # def exploration_strategy(self, unvisited_nodes:list) -> int:
    #     ''' Returns a node after sampling a probability distribution 
    #         of unvisited cities and their score 
    #         ========================================================
    #         score_array: applies the score_node function to all unvisited 
    #         nodes and returns an array.

    #     '''
        
    #     score_array = np.array(map(self.score_node, unvisited_nodes))
    #     sum_array = np.sum(score_array)
    #     prob_dist = score_array/sum_array
    #     return np.random.choice(a=unvisited_nodes, size=1, p=prob_dist)


    def choose_node(self, unvisited_nodes:list) -> int:
        ''' Chooses the next node to visit
        ================================================
        q: random value between 0,1
        q0: determines whether to explore or exploit
        score_array: applies the score_node function to all unvisited 
                     nodes and returns an array of results 
        Exploitation: chooses node based on the highest score
        Exploration: randomly chooses a node based on a probablity distribution 
        '''
        q = random.random() 
        unvisited_nodes = self.get_unvisited_nodes()
        
        score_array = np.array(map(self.score_node, unvisited_nodes))

        # Exploitation
        if q < self.q0:
            node = unvisited_nodes[np.argmax(score_array)]
            # node = self.exploitation_strategy(unvisited_nodes)    
        # Exploration
        elif q > self.q0:
            sum_array = np.sum(score_array)
            prob_dist = score_array/sum_array
            node = np.random.choice(a=unvisited_nodes, size=1, p=prob_dist)
            # node = self.exploration_strategy(unvisited_nodes)

        return node


        # return random.choice(unvisited_nodes)
    
    def visit_node(self, node:int) -> None:
        ''' Adds the node to the visited_nodes list'''
        self.visited_nodes.append(node)
        self.tour.append((self.current_node, node))
        self.current_node = node


    def complete_tour(self) -> None:
        ''' Takes the last node visited makes an edge to the first node'''

        #  Takes the visited_nodes list and another list with the first 
        #   node as last element in a list and zips them together
        # return list(zip(self.visited_nodes, self.visited_nodes[1:] + self.visited_nodes[:1]))

        self.tour.append((self.tour[-1][1], self.tour[0][0])) 
    
    def run(self) -> list[tuple]:
        ''' Main loop'''
        
        # While there are nodes to visit choose and visit them
        while self.get_unvisited_nodes():
            chosen_node = self.choose_node(self.get_unvisited_nodes())
            self.visit_node(chosen_node)
        
        self.complete_tour()
        return self.tour
```
