```python
import numpy as np
import matplotlib.pyplot as plt
import networkx as nx
```

```python
# https://stevedower.id.au/research/oliver-30 -> tsp 30 city problem source file

with open('coordinates.txt', 'r') as f:
    lines = [(line.strip().split(', ')) for line in f]
    coords = [(int(l[0]), int(l[1])) for l in lines]


with open('shortest_path.txt', 'r') as f:
    short_path = [int(j) for j in [i.split(" ") for i in f][0] if j!=""]
```

```python
np.linalg.norm(np.array(coords[0]) - np.array(coords[10]))
```

```python
fig, ax = plt.subplots(figsize=(12.2, 9))

labels = np.arange(1,len(coords)+1)
for (x, y), l in zip(coords, labels):
    ax.scatter(x, y, edgecolors='blue', c='white', s=300)
    ax.annotate(l, (x-0.7, y-0.5))
plt.plot((coords[0][0], coords[10][0]), (coords[0][1], coords[10][1]))
plt.show()
```

```python
graph = nx.Graph()
```

```python
node_labels = np.arange(1,len(coords)+1)    
```

```python
graph.add_nodes_from(node_labels)
```

```python
_pos = {node:positions for node,positions in zip(node_labels, coords)}
```

```python
def short_path_edge(short_path_list):
    path_tuples = [(short_path_list[i], short_path_list[i+1]) for i in range(len(short_path_list)-1)]
    path_tuples.append((short_path_list[0], short_path_list[-1])) # get the first and last one
    return path_tuples    
```

```python
def get_distance(node1, node2):
    '''nodes should be tuples(x,y)'''
    return np.linalg.norm(np.array(node1) - np.array(node2))
```

```python
G = nx.Graph()
```

```python
points = [(1, 10), (8, 10), (10, 8), (7, 4), (3, 1)]
edges = [(0, 1, 10), (1, 2, 5), (2, 3, 25), (0, 3, 3), (3, 4, 8)]
```

```python
def add_edge_to_graph(G, e1, e2, w):
    G.add_edge(e1, e2, weight=w)
```

```python
for i in range(len(edges)):
    add_edge_to_graph(G, points[edges[i][0]], points[edges[i][1]], edges[i][2])
```

```python
pos = {point: point for point in points}
pos
```

```python
fig, ax = plt.subplots()
nx.draw(G, pos=pos, node_color='k', ax=ax)
nx.draw(G, pos=pos, node_size=1500, ax=ax)
nx.draw_networkx_labels(G, pos=pos)
labels = nx.get_edge_attributes(G, 'weight')
nx.draw_networkx_edge_labels(G, pos, edge_labels=labels, ax=ax)
plt.axis("on")
ax.set_xlim(0, 11)
ax.set_ylim(0,11)
ax.tick_params(left=True, bottom=True, labelleft=True, labelbottom=True)
plt.show()
```
