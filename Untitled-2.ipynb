```python
def short_path_edge(short_path_list):
    path_tuples = [(short_path_list[i], short_path_list[i+1]) for i in range(len(short_path_list)-1)]
    path_tuples.append((short_path_list[0], short_path_list[-1])) # get the first and last one
    return path_tuples    
```

```python
def create_mat(prod_list, cust_list):
    mat = np.zeros((len(prod_list), len(cust_list)))
    for i, p in enumerate(prods):
        for j, c in enumerate(custs):
            try:
                mat[i][j] = sc[p][c]['quantity']
            except KeyError:
                continue
            
    return mat
```

```python
create_mat(prods, custs)
```

```python
trans_costs =[]
for (n1, n2) in qtys:
    try:
        trans_costs.append(sc[n1][n2]['quantity']*sc[n1][n2]['transport'])
    except KeyError:
        trans_costs.append(0)
```

```python
trans_costs = np.array(trans_costs)
trans_costs
```

```python
qs+trans_costs
```

```python
for i, (keys, calc) in enumerate(zip(qtys.keys(), qs+trans_costs)):
    if 'C1' in keys:
        print(keys, '->', calc, 'C1', 'index:',i)
    else:
        print(keys, '->', calc)
```
